```python
import pandas as pd

data = pd.read_csv("dataset/final_coastal_dataset.csv")
data["Date"] = pd.to_datetime(data["Date"])
print(data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
corr = data.drop("Date", axis=1).corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.plot(data["Date"], data["Rainfall"], label="Rainfall")
plt.plot(data["Date"], data["Sea_Level"], label="Sea Level")

plt.legend()
plt.title("Rainfall vs Sea Level Over Time")
plt.xlabel("Date")
plt.ylabel("Value")

plt.show()
```
